```python
import matplotlib.pyplot as plt
import numpy as np
import random
import math
import matplotlib.pyplot as plt
```

```python
# D > W,N
cores = np.linspace(1, 32, 32)
W = 30   # fix this!
centroids = np.linspace(1, 10, 10)
D = 1000
```

```python
# PLOT FOR CENTROID COUNT ANALYSIS and CORE COUNT ANALYSIS
def sys_arr_cycles(W,C,D):
    return W+C+D
def MMwoM_cycles(W,C,D,cores):
    return np.ceil(W*C*D/cores)

Cores, Centroids = np.meshgrid(cores, centroids)
sys_arr = sys_arr_cycles(W, Centroids, D)
MMwoM = MMwoM_cycles(W, Centroids, D, Cores)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(Cores, Centroids, sys_arr, 50, cmap='Blues')
ax.contour3D(Cores, Centroids, MMwoM, 50, cmap='viridis')
ax.set_xlabel('# of Cores (T)')
ax.set_ylabel('# of Centroid \nParitions (C)')
ax.set_zlabel('# of Clk Cycles', fontsize=10)
ax.set_title('Systolic Array Scheme (Blues) vs CPU Scheme (Viridis)\nLAC-based MM')
ax.zaxis.set_tick_params(labelsize=8)
```

```python
# plot a trace of centroids = 3 and # of cores for large D
def sys_arr_cycles(W,C,D):
    return W+C+D
def MMwoM_cycles(W,C,D,cores):
    return np.ceil(W*C*D/cores)

cores = np.linspace(1, 32, 32)
# cores = 5
# centroids = np.linspace(1, 10, 10)

sys_arr_1 = sys_arr_cycles(W, 1, D)
MMwoM_1 = MMwoM_cycles(W, 1, D, cores)
sys_arr_3 = sys_arr_cycles(W, 3, D)
MMwoM_3 = MMwoM_cycles(W, 3, D, cores)
sys_arr_5 = sys_arr_cycles(W, 5, D)
MMwoM_5 = MMwoM_cycles(W, 5, D, cores)

sys_arr_1 = [sys_arr_1]*32
sys_arr_3 = [sys_arr_3]*32
sys_arr_5 = [sys_arr_5]*32

plt.plot(cores, sys_arr_1, label="Systolic Array LAC-based MM, 1 Centroid")
plt.plot(cores, sys_arr_3, label="Systolic Array LAC-based MM, 3 Centroid")
plt.plot(cores, sys_arr_5, label="Systolic Array LAC-based MM, 5 Centroid")
plt.plot(cores, MMwoM_1, label="CPU LUT-based MM, 1 Centroid")
plt.plot(cores, MMwoM_3, label="CPU LUT-based MM, 3 Centroid")
plt.plot(cores, MMwoM_5, label="CPU LAC-based MM, 5 Centroid")
plt.xlabel("# of Cores")
plt.ylabel("# of Clk Cycles")
plt.legend()
```
